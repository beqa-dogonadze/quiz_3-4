```python
import pandas as pd

# 1)
df = pd.read_csv("Earthquakes_database.csv")
df = df.drop_duplicates()  #remove duplicates
df = df.drop(columns = 'Depth Seismic Stations')  #remove specific column

df["Depth Error"] = df["Depth Error"].fillna(0)  # replace NaN values with 0
df['Type and Depth Error'] = df['Type'].astype(str) + " "+ df['Depth Error'].astype(str) #combine 2 column

df = df.drop(columns = 'Type')  #remove specific column
df = df.drop(columns = 'Depth Error')  #remove specific column


df
```

```python
#2)

# bins and labels for the categories
bins = [3, 4, 5, 6, 7, 8, 10]

labels = ['Minor', 'Light', 'Moderate', 'Strong', 'Major', 'Great']

# add new category for classification
df['Earthquake category'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, include_lowest=True)

df
```

```python
# 4) 

# Bar plots
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(figsize=(13, 7))

applicant_id = df['Date'].tail(10)
income = df["Magnitude"].tail(10)
ax.set_xlabel('Year', fontsize = 20, color = 'r')
ax.set_ylabel('Magnitude', fontsize = 20, color = 'r')

ax.set_title('Last 10 earthquake and their magnitude ', fontsize = 25)
ax.bar(applicant_id, income, color = 'g', width = 0.3)

plt.show()

# Box plot
magn = df['Magnitude']
plt.boxplot(magn)
plt.show()


# histogram
labels = df['Date'].tail(10)
hist, edges, _ = plt.hist(labels, bins=10, color='blue', alpha=0.7)
plt.title('Magnitude Distribution')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')

plt.yticks([1,2,3,4,5,5,6,7,8])  # Adjust the range and step as needed

plt.show()
```
